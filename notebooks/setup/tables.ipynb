```sql
CREATE TABLE order_items (
    APP_NAME             VARCHAR(50),
    RESTAURANT_ID        VARCHAR(50),
    CREATION_TIME_UTC    DATETIME2,
    ORDER_ID             VARCHAR(50),
    USER_ID              VARCHAR(50)      NULL,
    PRINTED_CARD_NUMBER  VARCHAR(50)      NULL,
    IS_LOYALTY           VARCHAR(5),
    CURRENCY             VARCHAR(10),
    LINEITEM_ID          VARCHAR(50)      NOT NULL,
    ITEM_CATEGORY        VARCHAR(100),
    ITEM_NAME            VARCHAR(200),
    ITEM_PRICE           DECIMAL(10,2),
    ITEM_QUANTITY        INT);

CREATE TABLE order_item_options (
    ORDER_ID             VARCHAR(50),
    LINEITEM_ID          VARCHAR(50),
    OPTION_GROUP_NAME    VARCHAR(100),
    OPTION_NAME          VARCHAR(200),
    OPTION_PRICE         DECIMAL(10,2),
    OPTION_QUANTITY      INT
);

CREATE TABLE date_dim (
    date_key             VARCHAR(10),
    year                 INT,
    month                INT,
    week                 INT,
    day_of_week          VARCHAR(10),
    is_weekend           VARCHAR(5),
    is_holiday           VARCHAR(5),
    holiday_name         VARCHAR(100)     NULL);
```

```sql
use FOOD_APP_DB;
```

```sql
SELECT count(*) FROM order_items;
```

```sql
SELECT TOP 100 * FROM order_items;
```

```sql
SELECT count(*) FROM order_item_options;
```

```sql
SELECT TOP 100 * FROM order_item_options;
```
