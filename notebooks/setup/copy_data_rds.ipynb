```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_timestamp
import os

spark = SparkSession.builder \
    .appName("LoadOrderItems") \
    .config("spark.jars.packages", "com.microsoft.sqlserver:mssql-jdbc:12.6.1.jre11") \
    .getOrCreate()

HOST = "food-app-db.c0dkuk8u8kmc.us-east-1.rds.amazonaws.com"
PORT = 1433
DB   = "FOOD_APP_DB"
jdbc_url = f"jdbc:sqlserver://{HOST}:{PORT};databaseName={DB};encrypt=true;trustServerCertificate=true"
df = spark.read.option("header", True).option("multiLine", True).csv("order_items.csv")

df = df.filter(col("LINEITEM_ID").isNotNull() & (col("LINEITEM_ID") != ""))
df = df.withColumn("ITEM_PRICE", col("ITEM_PRICE").cast("decimal(10,2)")) \
       .withColumn("ITEM_QUANTITY", col("ITEM_QUANTITY").cast("int")) \
       .withColumn("CREATION_TIME_UTC", to_timestamp(col("CREATION_TIME_UTC")))

df.write \
  .format("jdbc") \
  .option("url", jdbc_url) \
  .option("dbtable", "order_items") \
  .option("user", "admin") \
  .option("password", os.environ["RDS_PASSWORD"]) \
  .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver") \
  .mode("append") \
  .save()
```
